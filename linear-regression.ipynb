```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate data along a line with random noise
x = np.linspace(0, 10, 100)
y = 2*x + np.random.normal(0, 1, 100)

plt.scatter(x,y)
```

```python
# Generate data along a line with random noise
m = np.linspace(0, 10, 100)
n = 2*x + np.random.normal(0, 5, 100)

plt.scatter(m,n)
```

```python
# Fit a linear regression model to the data
model = LinearRegression().fit(x.reshape(-1, 1), y.reshape(-1, 1))
model1 = LinearRegression().fit(m.reshape(-1, 1), n.reshape(-1, 1))
```

```python
# Evaluate the accuracy of the line
r2 = model.score(x.reshape(-1, 1), y.reshape(-1, 1))
print("R-squared value:", r2)
```

```python
# Evaluate the accuracy of the line
r21 = model1.score(m.reshape(-1, 1), n.reshape(-1, 1))
print("R-squared value:", r21)
```

```python
# Get the slope and intercept of the line of best fit
slope = model.coef_[0][0]
intercept = model.intercept_[0]

# Print the equation of the line of best fit
equation = f"y = {slope:.2f}x + {intercept:.2f}"
print("Equation of line:", equation)
```

```python
# Get the slope and intercept of the line of best fit
slope1 = model1.coef_[0][0]
intercept1 = model1.intercept_[0]

# Print the equation of the line of best fit
equation1 = f"n = {slope1:.2f}m + {intercept1:.2f}"
print("Equation of line:", equation1)
```

```python
# Plot the data and the line of best fit
plt.scatter(x, y)
plt.plot(x, model.predict(x.reshape(-1, 1)), color='red')
plt.title('Line of Best Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Plot the data and the line of best fit
plt.scatter(m, n)
plt.plot(m, model1.predict(m.reshape(-1, 1)), color='red')
plt.title('Line of Best Fit')
plt.xlabel('M')
plt.ylabel('N')
plt.show()
```

```python
# Plot a bar graph comparing the accuracy of the two lines of best fit
labels = ['Small Noise', 'Large Noise']
values = [r2, r21]
plt.bar(labels, values)
plt.title('R-Squared Values')
plt.xlabel('Noise Level')
plt.ylabel('R-Squared Value')
plt.show()
```
